```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

line plot

```python
x=[0.5,2,3,1,1.5,2,2.5,3,3.5]
y=[1,4,6,2,3,4,5,6,7]
# Resize your Graph (dpi specifies pixels per inch. When saving probably should use 300 if possible)
plt.figure(figsize=(12,6),dpi=300)
# Shorthand notation
plt.plot(x,y,'o--k',label='2x',linewidth=2,markersize=12)
# select interval we want to plot points at
x2=np.arange(0,20,2)
#plot a part of graph as dash line
plt.plot(x2[0:4],x2[0:4]*2,'o-g',label='x^2')
#plot a part of graph as double dash line
plt.plot(x2[3:10],x2[3:10]*2,'s--g',label='x^2')
# add title
plt.title('our first graph',fontdict={'fontname':'Comic sans MS','fontsize':20})
#add label to x axis
plt.xlabel('number in x')
#add label to y axis
plt.ylabel('number in y')
# X and Y tickmarks
plt.xticks([0,1,2,3,6,8])
plt.yticks([2,4,6,8])
# add a legend
plt.legend()
plt.show()
```

```python
fifa['Weight']=fifa.Weight.str.replace('lbs',' ').astype('float')
```

```python
fifa['Weight']
```

```python
light=fifa.loc[fifa['Weight']<125].count()[0]
medium_light=fifa.loc[(fifa['Weight']>=125) & (fifa['Weight']<150)].count()[0]
medium=fifa.loc[(fifa['Weight']>=150) & (fifa['Weight']<175)].count()[0]
medium_heavy=fifa.loc[(fifa['Weight']>=175) & (fifa['Weight']<200)].count()[0]
heavy=fifa.loc[fifa['Weight']>=200].count()[0]
heavy
```

```python
weigths= [light,medium_light, medium, medium_heavy, heavy]
plt.title('weight category pct of players')
plt.style.use('ggplot')
explode=(0.5,0.1,0,0,0.5)
labels=['UNDER 125','BETWEEN 125 AND 150','BETWEEN 150 AND 175', 'BETWEEN 175 AND 200', 'ABOVE 200']
plt.pie(weigths,labels=labels,autopct='%.2f%%',pctdistance=0.9,explode=explode)
plt.show()



```

```python
barcelona=fifa.loc[fifa['Club']=='FC Barcelona']['Overall']
realmadrid=fifa.loc[fifa['Club']=='Real Madrid']['Overall']
FCBayernMünchen=fifa.loc[fifa['Club']=='FC Bayern München']['Overall']
Milan=fifa.loc[fifa['Club']=='Milan']['Overall']
Liverpool=fifa.loc[fifa['Club']=='Liverpool']['Overall']
```

```python
plt.style.use('default')
plt.figure(figsize=(12,6))
plt.title('camparision of football clubs with respect to overall rated players ')
labels=['barcelona','realmadrid','FCBayernMünchen','Milan','Liverpool']
boxes=plt.boxplot([barcelona,realmadrid,FCBayernMünchen,Milan,Liverpool],labels=labels,patch_artist=True,medianprops={'linewidth':5})
for  box in boxes['boxes']:
    box.set(color='#a83c32', linewidth=4)


plt.xlabel('Clubs')
plt.ylabel('no of players in overallrating')
plt.show()
```
