### **`Problem-1:`** Draw a line plot of which, the x-axis is the "Year" and the y-axis is sum of "PM2.5" of two countries Iran and China.

```python
# code here
df = pd.read_csv('/content/List of most-polluted cities by particulate matter concentration - List of most-polluted cities by particulate matter concentration (2).csv')
df.head()
```

```python
df.columns
```

```python
x=df[(df['Country']=='China')].groupby('Year')['PM2.5'].sum()
```

```python
y=df[(df['Country']=='Iran')].groupby('Year')['PM2.5'].sum()
```

```python
plt.figure(figsize=(12,6))
plt.title('rate of increment in PM2.5 for iran and china over years', fontsize=14, fontweight='bold')
plt.plot(x.index,x.values,label='China',linestyle='--', color='blue', linewidth=2, marker='s',markersize=12)
plt.plot(y.index,y.values,label='Iran',linestyle='-', color='red', linewidth=2, marker='o',markersize=12)
plt.xlabel('Year')
plt.ylabel('PM2.5')
plt.style.use('seaborn-darkgrid')
plt.xticks(df['Year'].value_counts().index)
plt.grid()
plt.legend()
plt.show()
```

### **`Problem-2:`** Draw a histogram of the  column "PM10" of which the y-axis represents the probability (see the documentation how to draw the probability).

```python
plt.hist(df['PM10'],density=True,bins=50,facecolor='green',alpha=0.6)
plt.title('PM10 PDG')
plt.xlabel('PM10')
plt.ylabel('Probability')
plt.show()
```

### **`Problem-3:`** Draw a scatter plot where x-axis represents "PM2.5" and y-axis represents "PM10" for two countries Poland and Chile.

```python
plt.figure(figsize=(12,6),dpi=100)
plt.title('PM2.5 and PM10 comparison for poland and chile')
plt.scatter(df[df['Country']=='Poland']['PM2.5'],df[df['Country']=='Poland']['PM10'],marker='+',c='red',label='Poland')
plt.scatter(df[df['Country']=='Chile']['PM2.5'],df[df['Country']=='Chile']['PM10'],marker='o',c='black',label='chile')
plt.xlabel('PM2.5')
plt.ylabel('PM10')
plt.legend()
plt.grid()
plt.show()
```

```python
df1=df['Country'].value_counts().head(5)
```

### **`Problem-4:`** Draw a pie chart of top 5 most frequent countries.

```python
plt.figure(figsize=(12,6),dpi=100)
plt.title('top 5 most frequent countries')
plt.pie(df1,labels=df1.index,autopct='%.2f%%')
plt.show()
```

### **`Problem-5:`** Draw a bar chart which represents the counts of top 5 most frequent countries.



```python
plt.figure(figsize=(12,6),dpi=100)
plt.title('top 5 most frequent countries')
plt.bar(df1.index,df1)
plt.xlabel('Country')
plt.ylabel('frequency count')
plt.grid()
plt.show()
```

###`Problem-9` Show Quarter wise Sales data for all products as multi Bar chart.

```python
df.info()
```

```python
final_df = df.groupby(df['quarter']).sum()
```

```python
i=-1
for col in final_df.columns[1:7]:
  plt.bar(final_df.index+i,final_df[col],width=0.15,label=col)
  i=i-0.15
plt.xticks(final_df.index-1.4,final_df.index)
plt.xlabel('Product')
plt.ylabel('Sales')
plt.legend()
plt.show()
```
